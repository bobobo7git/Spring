# Spring Boot

> 스프링을 더욱 편리하게 사용할 수 있게 해준다.  
> 호환 가능한 라이브러리에 대한 버전을 가지고 있다.  

- 강의는 [spring.io](https://spring.io/projects/spring-boot#support) Spring Boot `3.3`버전 기준
    - 이 버전은 `Java 17`을 강제


### Spring Boot의 특징

- 독립 실행형 스프링 어플리케이션 생성
- `Tomcat`을 내장하고 있어, `WAR` 파일을 배포할 필요가 없다.
- `starter` 종속성을 제공하여 빌드 구성을 단순화
- 스프링, 서드파티 라이브러리의 버전을 자동으로 관리
- 상태 확인 기능 제공
  - CPU, 메모리 사용량 등 모니터링 가능(액튜에이터)
- `XML`에 대한 요구사항 없음
  - `Java Config` 방식 사용

## Spring Boot Project

- `Create New Spring Starter Project`

    
    - `Type`: Maven

    
    - `Spring Boot Version` 선택(`3.3`)
    - 의존성 설정
        - `Spring Boot DevTools`
        - `Spring Web`

### 프로젝트 실행

`src/main/resources/static`에 `index.html`을 만들어준다.
- Spring Boot에서는 기본적으로 `static/index.html`을 `Welcome Page`로 읽으려고 한다.

이후 프로젝트를 `Spring Boot App`으로 실행시켜준다.
- `src/main/java`의 `main` 함수가 실행된다.

    
    
실행 후 `Boot Dashboard`의 `지구본 모양`을 클릭하면 페이지가 열린다.
